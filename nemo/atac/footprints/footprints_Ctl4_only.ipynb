```r
data$Control.4.ap1.pos.cuts.bed=control_scale*data$Control.4.ap1.pos.cuts.bed
data$Control.4.ap4.pos.cuts.bed=control_scale*data$Control.4.ap4.pos.cuts.bed
data$Control.4.nfkb.pos.cuts.bed=control_scale*data$Control.4.nfkb.pos.cuts.bed
data$Control.4.runx1.pos.cuts.bed=control_scale*data$Control.4.runx1.pos.cuts.bed
data$Control.4.ctcf.pos.cuts.bed=control_scale*data$Control.4.ctcf.pos.cuts.bed
```

```r
data$Nemo.ap1.neg.cuts.bed=nemo_scale*data$Nemo.ap1.neg.cuts.bed
data$Nemo.ap4.neg.cuts.bed=nemo_scale*data$Nemo.ap4.neg.cuts.bed
data$Nemo.nfkb.neg.cuts.bed=nemo_scale*data$Nemo.nfkb.neg.cuts.bed
data$Nemo.runx1.neg.cuts.bed=nemo_scale*data$Nemo.runx1.neg.cuts.bed
data$Nemo.ctcf.neg.cuts.bed=nemo_scale*data$Nemo.ctcf.neg.cuts.bed
```

```r
data$Nemo.ap1.pos.cuts.bed=nemo_scale*data$Nemo.ap1.pos.cuts.bed
data$Nemo.ap4.pos.cuts.bed=nemo_scale*data$Nemo.ap4.pos.cuts.bed
data$Nemo.nfkb.pos.cuts.bed=nemo_scale*data$Nemo.nfkb.pos.cuts.bed
data$Nemo.runx1.pos.cuts.bed=nemo_scale*data$Nemo.runx1.pos.cuts.bed
data$Nemo.ctcf.pos.cuts.bed=nemo_scale*data$Nemo.ctcf.pos.cuts.bed
```

## AP1

```r
data$Control.4.ap1=data$Control.4.ap1.neg.cuts.bed+data$Control.4.ap1.pos.cuts.bed
data$Nemo.ap1=data$Nemo.ap1.neg.cuts.bed+data$Nemo.ap1.pos.cuts.bed

p1=ggplot(dat=data)+
geom_line(aes(x=data$Position,y=data$Control.4.ap1,color="AP1 Control "))+
geom_line(aes(x=data$Position,y=data$Nemo.ap1,color="AP1 NEMO"))+
xlab("Distance from Motif Center")+
ylab("Average Cuts")+
ggtitle("AP-1")+
scale_color_manual(values=c('#e41a1c','#377eb8'))+
theme_bw(20)

print(p1)
```
