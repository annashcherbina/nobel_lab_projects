```r
rm(list=ls())
library(ggplot2)
```

```r
data=read.table("summary.txt",header=TRUE,sep='\t')
```

```r
#annashch@nandi:/srv/scratch/annashch/nobel_lab_projects/nemo/atac/footprints$ wc -l *pos 
#  19021716 Control.tagAlign.pos
#  15652383 Nemo.tagAlign.pos
#  34674099 total

#annashch@nandi:/srv/scratch/annashch/nobel_lab_projects/nemo/atac/footprints$ wc -l *neg 
#  19021716 Control.tagAlign.neg
#  15652383 Nemo.tagAlign.neg
#  34674099 total

#scale factors 
nemo_scale=34674099/15652383
control_scale=34674099/19021716
data$Position=data$Position-100 
```

```r
data$Control.ap1.neg.cuts.bed=control_scale*data$Control.ap1.neg.cuts.bed
data$Control.ap4.neg.cuts.bed=control_scale*data$Control.ap4.neg.cuts.bed
data$Control.nfkb.neg.cuts.bed=control_scale*data$Control.nfkb.neg.cuts.bed
data$Control.runx1.neg.cuts.bed=control_scale*data$Control.runx1.neg.cuts.bed
data$Control.ctcf.neg.cuts.bed=control_scale*data$Control.ctcf.neg.cuts.bed
```

```r
data$Control.ap1.pos.cuts.bed=control_scale*data$Control.ap1.pos.cuts.bed
data$Control.ap4.pos.cuts.bed=control_scale*data$Control.ap4.pos.cuts.bed
data$Control.nfkb.pos.cuts.bed=control_scale*data$Control.nfkb.pos.cuts.bed
data$Control.runx1.pos.cuts.bed=control_scale*data$Control.runx1.pos.cuts.bed
data$Control.ctcf.pos.cuts.bed=control_scale*data$Control.ctcf.pos.cuts.bed
```

```r
data$Nemo.ap1.neg.cuts.bed=nemo_scale*data$Nemo.ap1.neg.cuts.bed
data$Nemo.ap4.neg.cuts.bed=nemo_scale*data$Nemo.ap4.neg.cuts.bed
data$Nemo.nfkb.neg.cuts.bed=nemo_scale*data$Nemo.nfkb.neg.cuts.bed
data$Nemo.runx1.neg.cuts.bed=nemo_scale*data$Nemo.runx1.neg.cuts.bed
data$Nemo.ctcf.neg.cuts.bed=nemo_scale*data$Nemo.ctcf.neg.cuts.bed
```

```r
data$Nemo.ap1.pos.cuts.bed=nemo_scale*data$Nemo.ap1.pos.cuts.bed
data$Nemo.ap4.pos.cuts.bed=nemo_scale*data$Nemo.ap4.pos.cuts.bed
data$Nemo.nfkb.pos.cuts.bed=nemo_scale*data$Nemo.nfkb.pos.cuts.bed
data$Nemo.runx1.pos.cuts.bed=nemo_scale*data$Nemo.runx1.pos.cuts.bed
data$Nemo.ctcf.pos.cuts.bed=nemo_scale*data$Nemo.ctcf.pos.cuts.bed
```

```r
head(data)
```

## AP1

```r
data$Control.ap1=data$Control.ap1.neg.cuts.bed+data$Control.ap1.pos.cuts.bed
data$Nemo.ap1=data$Nemo.ap1.neg.cuts.bed+data$Nemo.ap1.pos.cuts.bed

p1=ggplot(dat=data)+
geom_line(aes(x=data$Position,y=data$Control.ap1,color="AP1 Control "))+
geom_line(aes(x=data$Position,y=data$Nemo.ap1,color="AP1 NEMO"))+
xlab("Distance from Motif Center")+
ylab("Average Cuts")+
ggtitle("AP-1")+
scale_color_manual(values=c('#e41a1c','#377eb8'))+
theme_bw(20)

print(p1)
```
